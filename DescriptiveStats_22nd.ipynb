Women's Health Survey

In 1985, the USDA commissioned a study of women’s nutrition. Nutrient intake was measured for a random sample of 737 women aged 25-50 years. The following variables were measured:

- Calcium(mg)
-Iron(mg)
-Protein(g)
-Vitamin A(μg)
-Vitamin C(mg)


#Descriptve Statistics
We are going to anlyze this dataset on the basis of descriptive statistics functions that will help us in understanding the pattern and behavior of the data points.

```python
#Install the libraries
# We are using colab so no need to
```

```python
#Import all the important libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#pandas is used to read/load the dataset

data = pd.read_csv("nutrient.csv")
```

```python
data.head()
```

```python
print("Columns Name: ")
print(data.columns)

print("\nData Types: ")
print(data.dtypes)



```

```python
#dont depend data.describe(), you must CLEAN the data
data.describe()
```

```python
#Measure of Central Tendency

print("\nMean: ")
print(data.mean())

print("\nMedian: ")
print(data.median())

print("\nMode: ")
print(data.mode())
```

```python
#Measure of dispersion

#Variance--distance from mean value to other data points
#The square root of this is the standard deviation.
print("\nVariance: ")
print(data.var)


print("\nStandard Dveiation: ")
print(data.std())


print("\nRange: ")
print(data.max()-data.min())
```

```python
#Skewness and Kurtosis

print("\nSkewness: ")
print(data.skew())

print("\nKurtosis: ")
print(data.kurtosis())
```

```python
#Graph Analysis: Bar Chart

plt.figure(figsize=(6,4))  #size the graph
plt.bar(data['id'], data['calcium'], color = 'red', label='calcium') #define all data
plt.xlabel("ID")
plt.ylabel("Calcium")
plt.title("Calcium Intake by ID")
plt.legend()
plt.show()


```

```python
#Graph Analysis: Histogram

plt.figure(figsize=(6,4))
plt.hist(data['calcium'], bins=8, color = 'yellow', edgecolor='red') #define all data
plt.xlabel("Calcium")
plt.ylabel("ID")
plt.title("Histogram of Calcium Intake by ID")
plt.legend()
plt.show()
```
